**Section 1 - Loading and Cleaning COVID Data**

```python
import pandas as pd #load library and setting pd as the command name
path="https://pandemicdatalake.blob.core.windows.net/public/curated/covid-19/bing_covid-19_data/latest/bing_covid-19_data.parquet"
covid = pd.read_parquet(path) #read parquet file
covid.head(10) #shows top 10 results
```

```python
covid['Year'] = pd.DatetimeIndex(covid['updated']).year #Add a year column to dataframe
covid.iloc[0, covid.columns.get_loc("confirmed_change")] = 262
#locates the first index and replaces it with 262
covid
```

```python
covid1 = covid.rename(columns={"confirmed_change":"yearlycase", "country_region":"Country"})
 #rename columns
covid1
```

```python
import numpy as np
from numpy import nan as NA
covid1.dropna()
```

```python
covid2 = covid1.groupby(["Country", "Year"])[["yearlycase", "deaths", "recovered"]].sum().reset_index() #replaces old column index with new columns
covid2
```

**Section 3 - Training and Testing Data in Skicit Learn**



```python
from sklearn.model_selection import train_test_split
#the data is split between train and test
#training is used for building the model
#testing is used for creating the predictions
train,test=train_test_split(merged,test_size=0.25)  #25% data into test
test
```

```python
X_train = np.c_[train[['GDP']]]
y_train = np.c_[train[['yearlycase']]]
yt_train=np.reshape(y_train,len(train))   #Making row vector of y

X_test = np.c_[test[['GDP']]]
y_test = np.c_[test[['yearlycase']]]
yt_test=np.reshape(y_test,len(test))   #Making row vector of y

```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, yt_train)

from sklearn.metrics import mean_squared_error
y_predicted = lin_reg.predict(X_test)
MSE_lin_reg = mean_squared_error(yt_test, y_predicted)
RMSE_lin_reg = np.sqrt(MSE_lin_reg)
print("Mean Squared Error:", MSE_lin_reg)
print("Root Mean Squared Error:", RMSE_lin_reg)
print(np.c_[yt_test, y_predicted])
```

```python
 #artificial neural network
from sklearn.neural_network import MLPRegressor
neuralregressor = MLPRegressor()
neuralregressor.fit(X_train, yt_train)
from sklearn.metrics import mean_squared_error
y_predictedneural = neuralregressor.predict(X_test)
MSE_neuralregressor = mean_squared_error(yt_test, y_predictedneural)
RMSE_neuralregressor = np.sqrt(MSE_neuralregressor)
print("Mean Squared Error:" ,MSE_neuralregressor) #MSE produces a much smaller result
print("Root Mean Squared Error:", RMSE_neuralregressor)
print(np.c_[yt_test, y_predictedneural])

#result is closer with Linear Regression because of the datasets simplicity and the method's ability to capture relationships, making it more reliable
#There are not many factors that influence the Y variable
#result is farther with Neural Network. It is best suited for larger datasets but can also produce bias, making it
#unable to predict uncertain scenarios

```

```python
#Run Training Data

from pyspark.sql import SparkSession
#Create PySpark SparkSession
spark = SparkSession.builder \
    .master("local[1]") \
    .appName("SparkByExamples.com") \
    .getOrCreate()
#Create PySpark DataFrame from Pandas
sparktrain=spark.createDataFrame(train[['GDP', 'yearlycase']])
sparktrain.printSchema()
sparktrain.show()
```

```python
#Test Data

from pyspark.sql import SparkSession
#Create PySpark SparkSession
spark = SparkSession.builder \
    .master("local[1]") \
    .appName("SparkByExamples.com") \
    .getOrCreate()
#Create PySpark DataFrame from Pandas
sparktest=spark.createDataFrame(test[['GDP', 'yearlycase']]) #change of name but for test data
sparktest.printSchema()
sparktest.show()
```

```python
from pyspark.sql import Row
from pyspark.ml.linalg import Vectors

def transData(data):
    return data.rdd.map(lambda r: [Vectors.dense(r[:-1]),r[-1]]).toDF(['features','label'])

    #transform train and test to run ML
    #set as features and label
```

```python
transformedtrain= transData(sparktrain) #transform train data
transformedtrain.show(20) #features and label
#x(GDP) is features and Y(yearlycase) is label
#features can contain multiple x variables to influence the Y
```

```python
transformedtest= transData(sparktest) #transform test
transformedtest.show(20)
```

```python
from pyspark.ml.feature import VectorIndexer
# Automatically identify categorical features, and index them.
featureIndexer = VectorIndexer(inputCol="features", \
outputCol="indexedFeatures",\
maxCategories=4).fit(transformedtest)
```

```python
#To run linear regression, you have to import LinearRegression package from pyspark.ml.regression package
from pyspark.ml.regression import LinearRegression
#To perform multiple operations, Pipeline package needs to be imported
from pyspark.ml import Pipeline
# Define LinearRegression algorithm
lr = LinearRegression()  #Setting linear regression function
pipeline = Pipeline(stages=[featureIndexer, lr]) #As you see, featureIndexer is needed, een if you don't have categorical variable in X
model = pipeline.fit(transformedtrain)  #Train the model using tansformed train data

#using train data to run the model
#model has been built through the train data
#x and y needed to train the model
```

```python
predictions = model.transform(transformedtest)
# Select 5 rows to display
predictions.select("features","label","prediction").show(5)
#prediction figures are higher compared to actuals
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
# Select (prediction, true label) and compute test error
evaluator_mse = RegressionEvaluator(labelCol="label",
predictionCol="prediction",
metricName="mse")
mse = evaluator_mse.evaluate(predictions)
print("Mean Squared Error (MSE) on test data = %g" % mse)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
# Select (prediction, true label) and compute test error
evaluator_rmse = RegressionEvaluator(labelCol="label",
predictionCol="prediction",
metricName="rmse")
rmse = evaluator_rmse.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)

#MSE and RMSE are the same from Skicit Learn because data was simply converted with no actual change
```

```python
import matplotlib.pyplot as plt
plt.plot(test['GDP'], test['yearlycase'], 'o',color='black')
#plotting scatter plot of predicted y against x

plt.plot(test['GDP'], predictions.select("prediction").toPandas(), color='red')
plt.ylabel('yearlycase')
plt.xlabel('GDP')
#linear regression is basic form of prediction
#prediction comes from pyspark dataframe
#captured data with line of best fit, countries with high GDP did not hugely impact the total cases due to pauses in economic and social activity
```

```python
# Train a DecisionTree model.

from pyspark.ml.regression import DecisionTreeRegressor
decisiontree = DecisionTreeRegressor(featuresCol="indexedFeatures")
#Pipeline Architecture
# Chain indexer and tree in a Pipeline
pipeline = Pipeline(stages=[featureIndexer, decisiontree])
model_dt = pipeline.fit(transformedtrain)
```

```python
# Make predictions.
predictions_decisiontree = model_dt.transform(transformedtest)
predictions_decisiontree.select("features","label","prediction").show(5)
#shows top 5 rows
#decision tree has prediction closer to actuals
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
# Select (prediction, true label) and compute test error
evaluator_mse = RegressionEvaluator(labelCol="label",predictionCol="prediction",metricName="mse")

mse_decisiontree = evaluator_mse.evaluate(predictions_decisiontree)

print("Mean Squared Error (MSE) on test data = %g" % mse_decisiontree)
```
