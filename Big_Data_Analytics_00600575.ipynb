```python
from sklearn.model_selection import train_test_split
#the data is split between train and test
#training is used for building the model
#testing is used for creating the predictions
train,test=train_test_split(merged,test_size=0.25)  #25% data into test
test
```

```python
X_train = np.c_[train[['GDP']]]
y_train = np.c_[train[['yearlycase']]]
yt_train=np.reshape(y_train,len(train))   #Making row vector of y

#replace nan values with Numpy commands not Pandas

X_test = np.c_[test[['GDP']]]
y_test = np.c_[test[['yearlycase']]]
yt_test=np.reshape(y_test,len(test))   #Making row vector of y
```

```python
from sklearn.linear_model import LinearRegression
#linear regression creates a straight line of best fit
lin_reg = LinearRegression()
lin_reg.fit(X_train, yt_train)

from sklearn.metrics import mean_squared_error
y_predicted = lin_reg.predict(X_test)
MSE_lin_reg = mean_squared_error(yt_test, y_predicted)
RMSE_lin_reg = np.sqrt(MSE_lin_reg)
print("Mean Squared Error:", MSE_lin_reg)
print("Root Mean Squared Error:", RMSE_lin_reg)
print(np.c_[yt_test, y_predicted])
```

```python
from sklearn.neural_network import MLPRegressor #artificial neural network
neuralregressor = MLPRegressor()
neuralregressor.fit(X_train, yt_train)
from sklearn.metrics import mean_squared_error
y_predictedneural = neuralregressor.predict(X_test)
MSE_neuralregressor = mean_squared_error(yt_test, y_predictedneural)
RMSE_neuralregressor = np.sqrt(MSE_neuralregressor)
print("Mean Squared Error:" ,MSE_neuralregressor) #MSE produces a much smaller result
print("Root Mean Squared Error:", RMSE_neuralregressor)
print(np.c_[yt_test, y_predictedneural])

#result is closer with Neural Network most likely due to non-linearity and adaptability of this method.
#Therefore, it is more accurate and reliable.
#Since linear regression fits data linearly, it assumes correlations and relationships and it creates errors in data.
```

```python
#Run Training Data
from pyspark.sql import SparkSession
#Create PySpark SparkSession
spark = SparkSession.builder \
    .master("local[1]") \
    .appName("SparkByExamples.com") \
    .getOrCreate()
#Create PySpark DataFrame from Pandas
sparktrain=spark.createDataFrame(train[['GDP', 'yearlycase']])
sparktrain.printSchema()
sparktrain.show()
```

```python
#Test Data

from pyspark.sql import SparkSession
#Create PySpark SparkSession
spark = SparkSession.builder \
    .master("local[1]") \
    .appName("SparkByExamples.com") \
    .getOrCreate()
#Create PySpark DataFrame from Pandas
sparktest=spark.createDataFrame(test[['GDP', 'yearlycase']]) #change of name but for test data
sparktest.printSchema()
sparktest.show()
```

```python
from pyspark.sql import Row
from pyspark.ml.linalg import Vectors

def transData(data):
    return data.rdd.map(lambda r: [Vectors.dense(r[:-1]),r[-1]]).toDF(['features','label'])

    #transform train and test to run ML
```

```python
transformedtrain= transData(sparktrain) #transform train data
transformedtrain.show(20) #features and label
#all x and y is in features and label
#you can have multiple x's
```

```python
transformedtest= transData(sparktest) #transform test
transformedtest.show(20)
```

```python
from pyspark.ml.feature import VectorIndexer
# Automatically identify categorical features, and index them.
# We specify maxCategories so features with > 4 distinct values are treated as continuous.
featureIndexer = VectorIndexer(inputCol="features", \
outputCol="indexedFeatures",\
maxCategories=4).fit(transformedtest)
#If you want to see transformed data with indexedfeatures
#needed for ML

#using the test data to predict the Y variable for future data
#compare actual Y and predicted Y through MSE and RMSE
#using only x variable
```

```python
#To run linear regression, you have to import LinearRegression package from pyspark.ml.regression package
from pyspark.ml.regression import LinearRegression
#To perform multiple operations, Pipeline package needs to be imported
from pyspark.ml import Pipeline
# Define LinearRegression algorithm
lr = LinearRegression()  #Setting linear regression function
pipeline = Pipeline(stages=[featureIndexer, lr]) #As you see, featureIndexer is needed, een if you don't have categorical variable in X
model = pipeline.fit(transformedtrain)  #Train the model using tansformed train data

#using train data to run the model
#model has been built through the train data
#x and y needed to train the model
```

```python
predictions = model.transform(transformedtest)
# Select 5 rows to display
predictions.select("features","label","prediction").show(5)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
# Select (prediction, true label) and compute test error
evaluator_mse = RegressionEvaluator(labelCol="label",
predictionCol="prediction",
metricName="mse")
mse = evaluator_mse.evaluate(predictions)
print("Mean Squared Error (MSE) on test data = %g" % mse)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
# Select (prediction, true label) and compute test error
evaluator_rmse = RegressionEvaluator(labelCol="label",
predictionCol="prediction",
metricName="rmse")
rmse = evaluator_rmse.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)

#RMSE and SME are the same values from Skicit Learn
```
